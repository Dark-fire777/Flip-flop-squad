```python
"""
ADVANCED Semiconductor Defect Classification
Realistic dataset with proper defect simulation
Building on 92% baseline to push higher!
"""

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import cv2
import os
from pathlib import Path
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from scipy.ndimage import gaussian_filter
import albumentations as A

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

class Config:
    IMG_SIZE = 128  # Higher resolution for better features
    CHANNELS = 1
    NUM_CLASSES = 8

    BATCH_SIZE = 32
    EPOCHS = 50
    INITIAL_LR = 0.001

    # More samples for better learning
    SAMPLES_PER_CLASS = 800  # 6,400 total images

    DATASET_DIR = 'advanced_defects'
    OUTPUT_DIR = 'advanced_outputs'

    CLASS_NAMES = [
        'good_die', 'particle_defect', 'scratch_defect', 'spot_defect',
        'edge_chip', 'stain_defect', 'crack_defect', 'residue_defect'
    ]

os.makedirs(f'{Config.OUTPUT_DIR}/models', exist_ok=True)
os.makedirs(f'{Config.OUTPUT_DIR}/results', exist_ok=True)
os.makedirs(f'{Config.OUTPUT_DIR}/samples', exist_ok=True)

class RealisticDefectGenerator:
    """Generate realistic semiconductor defects"""

    def __init__(self, img_size=128):
        self.size = img_size

    def create_wafer_texture(self):
        """Create realistic silicon wafer background"""
        img = np.ones((self.size, self.size), dtype=np.float32) * 0.5

        # Add silicon crystal structure texture
        for _ in range(50):
            x, y = np.random.randint(0, self.size, 2)
            intensity = np.random.uniform(0.4, 0.6)
            size = np.random.randint(3, 8)
            cv2.circle(img, (x, y), size, intensity, -1)

        img = gaussian_filter(img, sigma=1.5)

        # Add subtle grain
        grain = np.random.normal(0, 0.02, img.shape)
        img = np.clip(img + grain, 0, 1)

        return img

    def add_die_pattern(self, img):
        """Add circuit die pattern"""
        # Grid lines
        grid_spacing = self.size // 8
        for i in range(0, self.size, grid_spacing):
            thickness = 1 if np.random.random() > 0.5 else 2
            cv2.line(img, (i, 0), (i, self.size), 0.7, thickness)
            cv2.line(img, (0, i), (self.size, i), 0.7, thickness)

        # Add circuit elements (rectangles, small patterns)
        for _ in range(20):
            x = np.random.randint(5, self.size - 15)
            y = np.random.randint(5, self.size - 15)
            w = np.random.randint(3, 10)
            h = np.random.randint(3, 10)
            cv2.rectangle(img, (x, y), (x + w, y + h), 0.65, -1)

        return img

    def create_good_die(self):
        """Perfect die with clean circuit pattern"""
        img = self.create_wafer_texture()
        img = self.add_die_pattern(img)

        # Very minimal noise
        noise = np.random.normal(0, 0.01, img.shape)
        img = np.clip(img + noise, 0, 1)

        return img

    def create_particle_defect(self):
        """Contamination particles on surface"""
        img = self.create_wafer_texture()
        img = self.add_die_pattern(img)

        # Add 2-5 particles of varying sizes
        num_particles = np.random.randint(2, 6)
        for _ in range(num_particles):
            x = np.random.randint(10, self.size - 10)
            y = np.random.randint(10, self.size - 10)
            radius = np.random.randint(3, 12)

            # Bright particle with soft edge
            particle = np.zeros_like(img)
            cv2.circle(particle, (x, y), radius, 1.0, -1)
            particle = gaussian_filter(particle, sigma=1.0)

            # Blend particle onto image
            img = np.clip(img + particle * 0.4, 0, 1)

        noise = np.random.normal(0, 0.015, img.shape)
        img = np.clip(img + noise, 0, 1)

        return img

    def create_scratch_defect(self):
        """Linear scratch across die"""
        img = self.create_wafer_texture()
        img = self.add_die_pattern(img)

        # Create realistic scratch
        num_scratches = np.random.randint(1, 3)
        for _ in range(num_scratches):
            # Random start and end points
            x1 = np.random.randint(0, self.size)
            y1 = np.random.randint(0, self.size // 4)
            x2 = np.random.randint(0, self.size)
            y2 = np.random.randint(3 * self.size // 4, self.size)

            # Draw dark scratch with varying width
            thickness = np.random.randint(2, 5)
            cv2.line(img, (x1, y1), (x2, y2), 0.15, thickness)

            # Add rough edges
            num_points = 20
            for i in range(num_points):
                t = i / num_points
                x = int(x1 + t * (x2 - x1))
                y = int(y1 + t * (y2 - y1))
                offset = np.random.randint(-2, 3)
                cv2.circle(img, (x + offset, y + offset), 1, 0.1, -1)

        noise = np.random.normal(0, 0.015, img.shape)
        img = np.clip(img + noise, 0, 1)

        return img

    def create_spot_defect(self):
        """Dark spots from contamination"""
        img = self.create_wafer_texture()
        img = self.add_die_pattern(img)

        # Add 3-8 dark spots
        num_spots = np.random.randint(3, 9)
        for _ in range(num_spots):
            x = np.random.randint(10, self.size - 10)
            y = np.random.randint(10, self.size - 10)
            radius = np.random.randint(4, 10)

            # Dark spot with gradient
            spot = np.zeros_like(img)
            cv2.circle(spot, (x, y), radius, 1.0, -1)
            spot = gaussian_filter(spot, sigma=1.5)

            img = np.clip(img - spot * 0.35, 0, 1)

        noise = np.random.normal(0, 0.015, img.shape)
        img = np.clip(img + noise, 0, 1)

        return img

    def create_edge_chip(self):
        """Chipped or broken edge"""
        img = self.create_wafer_texture()
        img = self.add_die_pattern(img)

        # Create chip in corner or edge
        edge = np.random.choice(['corner', 'edge'])

        if edge == 'corner':
            # Random corner
            corner = np.random.randint(0, 4)
            size = np.random.randint(self.size // 4, self.size // 2)

            if corner == 0:  # Top-left
                img[0:size, 0:size] = 0
            elif corner == 1:  # Top-right
                img[0:size, -size:] = 0
            elif corner == 2:  # Bottom-left
                img[-size:, 0:size] = 0
            else:  # Bottom-right
                img[-size:, -size:] = 0

            # Irregular edge
            for i in range(size):
                offset = np.random.randint(-5, 6)
                if corner == 0:
                    img[i, size:size + offset] = 0.2
                elif corner == 1:
                    img[i, -size - offset:-size] = 0.2
        else:
            # Edge chip
            edge_side = np.random.randint(0, 4)
            width = np.random.randint(self.size // 6, self.size // 3)

            if edge_side == 0:  # Top
                img[0:width, :] = 0
            elif edge_side == 1:  # Bottom
                img[-width:, :] = 0
            elif edge_side == 2:  # Left
                img[:, 0:width] = 0
            else:  # Right
                img[:, -width:] = 0

        noise = np.random.normal(0, 0.015, img.shape)
        img = np.clip(img + noise, 0, 1)

        return img

    def create_stain_defect(self):
        """Stain or residue blob"""
        img = self.create_wafer_texture()
        img = self.add_die_pattern(img)

        # Create 1-2 irregular stains
        num_stains = np.random.randint(1, 3)
        for _ in range(num_stains):
            cx = np.random.randint(self.size // 4, 3 * self.size // 4)
            cy = np.random.randint(self.size // 4, 3 * self.size // 4)

            # Create irregular stain shape
            stain = np.zeros_like(img)
            for _ in range(50):
                angle = np.random.uniform(0, 2 * np.pi)
                radius = np.random.randint(10, 25)
                x = int(cx + radius * np.cos(angle))
                y = int(cy + radius * np.sin(angle))
                size = np.random.randint(4, 10)
                cv2.circle(stain, (x, y), size, 1.0, -1)

            stain = gaussian_filter(stain, sigma=3.0)
            img = np.clip(img - stain * 0.25, 0, 1)

        noise = np.random.normal(0, 0.015, img.shape)
        img = np.clip(img + noise, 0, 1)

        return img

    def create_crack_defect(self):
        """Crack through die"""
        img = self.create_wafer_texture()
        img = self.add_die_pattern(img)

        # Create branching crack
        start_x = np.random.randint(0, self.size)
        start_y = 0

        x, y = start_x, start_y
        main_crack = [(x, y)]

        # Main crack path
        for _ in range(15):
            x += np.random.randint(-8, 9)
            y += np.random.randint(6, 12)
            x = np.clip(x, 0, self.size - 1)
            y = np.clip(y, 0, self.size - 1)
            main_crack.append((x, y))
            if y >= self.size - 1:
                break

        # Draw main crack
        for i in range(len(main_crack) - 1):
            cv2.line(img, main_crack[i], main_crack[i + 1], 0.05, 2)

        # Add branches
        for i in range(2, len(main_crack) - 2, 3):
            branch_len = np.random.randint(3, 8)
            bx, by = main_crack[i]
            for _ in range(branch_len):
                bx += np.random.randint(-5, 6)
                by += np.random.randint(-3, 4)
                cv2.circle(img, (bx, by), 1, 0.1, -1)

        noise = np.random.normal(0, 0.015, img.shape)
        img = np.clip(img + noise, 0, 1)

        return img

    def create_residue_defect(self):
        """Chemical residue spots"""
        img = self.create_wafer_texture()
        img = self.add_die_pattern(img)

        # Add multiple bright residue spots
        num_residues = np.random.randint(5, 15)
        for _ in range(num_residues):
            x = np.random.randint(5, self.size - 5)
            y = np.random.randint(5, self.size - 5)
            radius = np.random.randint(2, 7)

            # Bright residue with halo
            residue = np.zeros_like(img)
            cv2.circle(residue, (x, y), radius, 1.0, -1)
            cv2.circle(residue, (x, y), radius + 2, 0.5, 1)
            residue = gaussian_filter(residue, sigma=0.8)

            img = np.clip(img + residue * 0.35, 0, 1)

        noise = np.random.normal(0, 0.015, img.shape)
        img = np.clip(img + noise, 0, 1)

        return img

def create_advanced_dataset():
    """Create realistic semiconductor defect dataset"""
    print(f"\n{'='*70}")
    print("Creating ADVANCED Realistic Dataset")
    print(f"{'='*70}\n")

    generator = RealisticDefectGenerator(Config.IMG_SIZE)

    # Create sample visualization
    fig, axes = plt.subplots(2, 4, figsize=(16, 8))
    axes = axes.flatten()

    defect_funcs = [
        generator.create_good_die,
        generator.create_particle_defect,
        generator.create_scratch_defect,
        generator.create_spot_defect,
        generator.create_edge_chip,
        generator.create_stain_defect,
        generator.create_crack_defect,
        generator.create_residue_defect
    ]

    for class_idx, class_name in enumerate(Config.CLASS_NAMES):
        class_dir = f"{Config.DATASET_DIR}/{class_name}"
        os.makedirs(class_dir, exist_ok=True)

        print(f"Generating {class_name}...")

        for i in tqdm(range(Config.SAMPLES_PER_CLASS), desc=f"  {class_name}"):
            # Generate defect
            img = defect_funcs[class_idx]()

            # Convert to uint8
            img_uint8 = (img * 255).astype(np.uint8)

            # Save
            cv2.imwrite(f"{class_dir}/{class_name}_{i:04d}.png", img_uint8)

            # Save first sample for visualization
            if i == 0:
                axes[class_idx].imshow(img, cmap='gray', vmin=0, vmax=1)
                axes[class_idx].set_title(class_name.replace('_', ' ').title())
                axes[class_idx].axis('off')

    plt.tight_layout()
    plt.savefig(f"{Config.OUTPUT_DIR}/samples/dataset_samples.png", dpi=200, bbox_inches='tight')
    print(f"\n✓ Dataset samples saved to {Config.OUTPUT_DIR}/samples/")

    total_images = Config.NUM_CLASSES * Config.SAMPLES_PER_CLASS
    print(f"\n✓ Created {total_images:,} realistic images")
    print(f"{'='*70}\n")

def create_augmentation():
    """Data augmentation pipeline"""
    return A.Compose([
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        A.Rotate(limit=15, p=0.5),
        A.RandomBrightnessContrast(
            brightness_limit=0.1,
            contrast_limit=0.1,
            p=0.5
        ),
        A.GaussNoise(var_limit=(0.001, 0.005), p=0.3),
    ])

def load_data_with_augmentation():
    """Load data with augmentation"""
    print(f"\n{'='*70}")
    print("Loading Data with Augmentation")
    print(f"{'='*70}\n")

    augmentation = create_augmentation()

    X_all = []
    y_all = []

    for class_idx, class_name in enumerate(Config.CLASS_NAMES):
        class_dir = f"{Config.DATASET_DIR}/{class_name}"
        images = sorted([f for f in os.listdir(class_dir) if f.endswith('.png')])

        print(f"Loading {class_name}: {len(images)} images")

        for img_file in images:
            img = cv2.imread(os.path.join(class_dir, img_file), cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (Config.IMG_SIZE, Config.IMG_SIZE))

            # Apply augmentation 50% of the time during loading
            if np.random.random() < 0.5:
                augmented = augmentation(image=img)
                img = augmented['image']

            img = img.astype(np.float32) / 255.0

            X_all.append(img)
            y_all.append(class_idx)

    X_all = np.array(X_all)[..., np.newaxis]
    y_all = np.array(y_all)

    # Shuffle
    indices = np.random.permutation(len(X_all))
    X_all = X_all[indices]
    y_all = y_all[indices]

    # Split: 70% train, 15% val, 15% test
    n_total = len(X_all)
    n_train = int(n_total * 0.70)
    n_val = int(n_total * 0.15)

    X_train = X_all[:n_train]
    y_train = y_all[:n_train]

    X_val = X_all[n_train:n_train + n_val]
    y_val = y_all[n_train:n_train + n_val]

    X_test = X_all[n_train + n_val:]
    y_test = y_all[n_train + n_val:]

    # Convert to categorical
    y_train = keras.utils.to_categorical(y_train, Config.NUM_CLASSES)
    y_val = keras.utils.to_categorical(y_val, Config.NUM_CLASSES)
    y_test = keras.utils.to_categorical(y_test, Config.NUM_CLASSES)

    print(f"\nTrain: {len(X_train):,}")
    print(f"Val:   {len(X_val):,}")
    print(f"Test:  {len(X_test):,}")
    print(f"{'='*70}\n")

    return (X_train, y_train), (X_val, y_val), (X_test, y_test)

def build_advanced_model():
    """Build improved CNN model"""
    print(f"\n{'='*70}")
    print("Building Advanced Model")
    print(f"{'='*70}\n")

    model = keras.Sequential([
        layers.Input(shape=(Config.IMG_SIZE, Config.IMG_SIZE, 1)),

        # Block 1
        layers.Conv2D(32, 3, padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.Conv2D(32, 3, padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D(2),
        layers.Dropout(0.25),

        # Block 2
        layers.Conv2D(64, 3, padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.Conv2D(64, 3, padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D(2),
        layers.Dropout(0.25),

        # Block 3
        layers.Conv2D(128, 3, padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.Conv2D(128, 3, padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D(2),
        layers.Dropout(0.25),

        # Block 4
        layers.Conv2D(256, 3, padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.GlobalAveragePooling2D(),

        # Classifier
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(Config.NUM_CLASSES, activation='softmax')
    ])

    model.summary()
    print(f"\nTotal Parameters: {model.count_params():,}")
    print(f"{'='*70}\n")

    return model

def train_advanced_model(model, train_data, val_data):
    """Train with advanced callbacks"""
    print(f"\n{'='*70}")
    print("Training Advanced Model")
    print(f"{'='*70}\n")

    X_train, y_train = train_data
    X_val, y_val = val_data

    model.compile(
        optimizer=keras.optimizers.Adam(Config.INITIAL_LR),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    callbacks = [
        keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=15,
            restore_best_weights=True,
            verbose=1
        ),
        keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            min_lr=1e-7,
            verbose=1
        ),
        keras.callbacks.ModelCheckpoint(
            f"{Config.OUTPUT_DIR}/models/best_model.keras",
            monitor='val_accuracy',
            save_best_only=True,
            verbose=1
        )
    ]

    history = model.fit(
        X_train, y_train,
        batch_size=Config.BATCH_SIZE,
        epochs=Config.EPOCHS,
        validation_data=(X_val, y_val),
        callbacks=callbacks,
        verbose=1
    )

    # Plot training history
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    # Accuracy
    ax1.plot(history.history['accuracy'], label='Train')
    ax1.plot(history.history['val_accuracy'], label='Validation')
    ax1.set_title('Model Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Loss
    ax2.plot(history.history['loss'], label='Train')
    ax2.plot(history.history['val_loss'], label='Validation')
    ax2.set_title('Model Loss')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(f"{Config.OUTPUT_DIR}/results/training_history.png", dpi=200)
    print(f"\n✓ Training history saved")

    return history

def evaluate_advanced_model(model, test_data):
    """Comprehensive evaluation"""
    print(f"\n{'='*70}")
    print("Evaluating Advanced Model")
    print(f"{'='*70}\n")

    X_test, y_test = test_data

    # Evaluate
    loss, acc = model.evaluate(X_test, y_test, verbose=0)

    # Predictions
    y_pred = model.predict(X_test, verbose=0)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(y_test, axis=1)

    # Confusion matrix
    cm = confusion_matrix(y_true_classes, y_pred_classes)

    print(f"{'='*70}")
    print(f"FINAL TEST RESULTS")
    print(f"{'='*70}")
    print(f"Test Accuracy: {acc*100:.2f}%")
    print(f"Test Loss:     {loss:.4f}")
    print(f"{'='*70}\n")

    # Classification report
    print("Detailed Classification Report:")
    print("="*70)
    report = classification_report(
        y_true_classes, y_pred_classes,
        target_names=Config.CLASS_NAMES,
        digits=4
    )
    print(report)

    # Per-class accuracy
    print("\nPer-Class Accuracy:")
    print("="*70)
    for i, class_name in enumerate(Config.CLASS_NAMES):
        class_acc = cm[i, i] / cm[i, :].sum() * 100
        print(f"{class_name:20s}: {class_acc:6.2f}%")

    # Plot confusion matrix
    plt.figure(figsize=(12, 10))
    sns.heatmap(
        cm, annot=True, fmt='d', cmap='Blues',
        xticklabels=[name.replace('_', '\n') for name in Config.CLASS_NAMES],
        yticklabels=[name.replace('_', '\n') for name in Config.CLASS_NAMES],
        cbar_kws={'label': 'Count'}
    )
    plt.title(f'Confusion Matrix - Test Accuracy: {acc*100:.2f}%', fontsize=14, pad=20)
    plt.ylabel('True Label', fontsize=12)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.tight_layout()
    plt.savefig(f"{Config.OUTPUT_DIR}/results/confusion_matrix.png", dpi=200, bbox_inches='tight')
    print(f"\n✓ Confusion matrix saved")

    # Save final model
    model.save(f"{Config.OUTPUT_DIR}/models/final_advanced_model.keras")
    print(f"✓ Model saved: {Config.OUTPUT_DIR}/models/final_advanced_model.keras\n")

    return acc

def test_live_predictions(model, test_data):
    """Test on random samples with confidence scores"""
    X_test, y_test = test_data

    # Pick 20 random samples
    indices = np.random.choice(len(X_test), 20, replace=False)

    print(f"\n{'='*70}")
    print("LIVE TESTING - 20 Random Samples")
    print(f"{'='*70}\n")

    correct = 0
    for i, idx in enumerate(indices):
        img = X_test[idx:idx+1]
        true_class = np.argmax(y_test[idx])

        pred = model.predict(img, verbose=0)[0]
        pred_class = np.argmax(pred)
        confidence = pred[pred_class] * 100

        is_correct = (pred_class == true_class)
        if is_correct:
            correct += 1

        status = "✓" if is_correct else "✗"
        true_name = Config.CLASS_NAMES[true_class]
        pred_name = Config.CLASS_NAMES[pred_class]

        print(f"{i+1:2d}. True: {true_name:20s} | "
              f"Pred: {pred_name:20s} | "
              f"Conf: {confidence:5.1f}% {status}")

    accuracy = correct / 20 * 100
    print(f"\n{'='*70}")
    print(f"Live Test Accuracy: {correct}/20 = {accuracy:.1f}%")
    print(f"{'='*70}\n")

def main():
    print(f"\n{'#'*70}")
    print("# ADVANCED Semiconductor Defect Classification")
    print("# Realistic Dataset - Building on 92% baseline")
    print(f"{'#'*70}\n")

    np.random.seed(42)
    tf.random.set_seed(42)

    # Create advanced dataset
    if not os.path.exists(Config.DATASET_DIR):
        create_advanced_dataset()
    else:
        print(f"✓ Dataset exists at {Config.DATASET_DIR}\n")

    # Load data with augmentation
    train_data, val_data, test_data = load_data_with_augmentation()

    # Build advanced model
    model = build_advanced_model()

    # Train
    history = train_advanced_model(model, train_data, val_data)

    # Evaluate
    accuracy = evaluate_advanced_model(model, test_data)

    # Live testing
    test_live_predictions(model, test_data)

    # Final summary
    print(f"\n{'='*70}")
    print(f"TRAINING COMPLETE!")
    print(f"{'='*70}")
    print(f"Final Test Accuracy: {accuracy*100:.2f}%")
    print(f"Previous Baseline:   92.00%")

    improvement = accuracy * 100 - 92.0
    if improvement > 0:
        print(f"Improvement:         +{improvement:.2f}%")
    else:
        print(f"Change:              {improvement:.2f}%")

    print(f"\nModel saved: {Config.OUTPUT_DIR}/models/final_advanced_model.keras")
    print(f"Results:     {Config.OUTPUT_DIR}/results/")
    print(f"{'='*70}\n")

if __name__ == "__main__":
    main()
```
